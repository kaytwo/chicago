```python
import matplotlib.pyplot as plt
import geopandas
import pandas
```

```python
from spatial_ops import assign
```

```python
blocks = geopandas.read_file("./shapefiles/chicago_precincts_2012.shp")
vtds = geopandas.read_file("./shapefiles/chicago_community_areas.shp")
```

```python
assignment = assign(blocks, vtds)
weights = blocks.area / assignment.map(vtds.area)
prorated = pandas.Series(assignment.map(vtds["NH_BLACK"]) * weights)
```

```python
errors = prorated - blocks["NH_BLACK"]
pct_errors = errors[blocks["NH_BLACK"] > 0] / blocks["NH_BLACK"][blocks["NH_BLACK"] > 0]
```
